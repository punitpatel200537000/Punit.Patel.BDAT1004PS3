## <font color='blue'>Question 6: Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend. Don't forget to label your axes!</font>

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

data.set_index('Year', inplace=True)

plt.figure(figsize=(12, 6))
plt.plot(data['Marriages_per_1000'], label='Marriages')
plt.plot(data['Divorces_per_1000'], label='Divorces')

plt.xlabel('Year')
plt.ylabel('Number per 1000 people')
plt.title('Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.legend()

plt.show()
```

## <font color='blue'>Question 7: Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000. Don't forget to label your axes!</font>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV data
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')

# Filter the data for the years 1900, 1950, and 2000
data_filtered = data[data['Year'].isin([1900, 1950, 2000])]

# Set the index to the 'Year' column
data_filtered.set_index('Year', inplace=True)

# Create a vertical bar chart
ax = data_filtered[['Marriages_per_1000', 'Divorces_per_1000']].plot(kind='bar', figsize=(12, 6))

# Add labels
plt.xlabel('Year')
plt.ylabel('Number per 1000 people')
plt.title('Comparison of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')

# Customize the plot
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)

# Display the plot
plt.show()
```

## <font color='blue'>Question 9: Create a pie chart showing the fraction of all Roman Emperors that were assassinated.<br>Make sure that the pie chart is an even circle, labels the categories, and shows the percentage breakdown of the categories.</font>

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV data
data = pd.read_csv('roman-emperor-reigns.csv')

# Count the causes of death
cause_counts = data['Cause_of_Death'].value_counts()

# Filter only assassinations
assassinations = cause_counts[cause_counts.index.str.contains('Assassinated')].sum()
possibly_assassinated = cause_counts[cause_counts.index.str.contains('Possibly assassinated')].sum()

# Calculate the fraction of assassinated emperors
fraction_assassinated = assassinations / cause_counts.sum()
fraction_possibly_assassinated = possibly_assassinated / cause_counts.sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
labels = ['Assassinated', 'Possibly assassinated', 'Others']
sizes = [fraction_assassinated, fraction_possibly_assassinated, 1 - (fraction_assassinated + fraction_possibly_assassinated)]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Make sure the pie chart is an even circle
plt.axis('equal')

# Add a title
plt.title('Fraction of Roman Emperors Assassinated')

# Display the plot
plt.show()
```

## <font color='blue'>Question 10: Create a scatter plot showing the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between 2000 and 2009. <br>Don't forget to label your axes! <br>Color each dot according to its year.</font>

```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV data
data = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv')

# Create a scatter plot
plt.figure(figsize=(11, 7))
scatter = plt.scatter(data['Total Arcade Revenue (billions)'],
                      data['Computer Science Doctorates Awarded (US)'],
                      c=data['Year'], cmap='viridis')

# Add colorbar with years
cbar = plt.colorbar(scatter)
cbar.set_label('Year')

# Add labels
plt.xlabel('Total Arcade Revenue (in billions)')
plt.ylabel('Computer Science Doctorates Awarded (US)')
plt.title('Total Arcade Revenue vs Computer Science Doctorates Awarded (2000-2009)')

# Display the plot
plt.show()
```
